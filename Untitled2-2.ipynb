```python
# Sensitivity Curve Function
def plot_sensitivity_curve(y_true, y_pred_proba, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc(fpr, tpr):.2f})')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Sensitivity Curve')
    plt.legend()

# Individual Model Training and Evaluation
def evaluate_model(model, X_resampled, y_resampled, model_name):
    model.fit(X_resampled, y_resampled)
    y_pred_proba = model.predict_proba(X_resampled)[:, 1]
    f1 = f1_score(y_resampled, model.predict(X_resampled))
    roc_auc = roc_auc_score(y_resampled, y_pred_proba)
    precision, recall, _ = precision_recall_curve(y_resampled, y_pred_proba)
    plot_sensitivity_curve(y_resampled, y_pred_proba, model_name)
    print(f"{model_name} - F1 Score: {f1}, ROC-AUC Score: {roc_auc}")
    return {'F1 Score': f1, 'ROC-AUC': roc_auc}
```

```python
def plot_ssa_curve(y_true, y_pred_proba, model_name):
    thresholds = np.linspace(0, 1, 100)
    accuracy = []
    sensitivity = []
    specificity = []
    
    for t in thresholds:
        y_pred = (y_pred_proba >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
        
        acc = (tp + tn) / (tp + tn + fp + fn)
        sens = tp / (tp + fn) if (tp + fn) > 0 else 0
        spec = tn / (tn + fp) if (tn + fp) > 0 else 0
        
        accuracy.append(acc)
        sensitivity.append(sens)
        specificity.append(spec)
    
    cutoff_df = pd.DataFrame({'prob': thresholds, 'accuracy': accuracy, 'sensi': sensitivity, 'speci': specificity})
    
    plt.figure(figsize=(8, 5))
    cutoff_df.plot(x='prob', y=['accuracy', 'sensi', 'speci'])
    optimal_thresh = thresholds[np.argmax(np.array(sensitivity) + np.array(specificity) - 1)]  # Youden’s J statistic
    plt.vlines(x=optimal_thresh, ymin=0, ymax=1, colors='r', linestyles="--")
    plt.title(f"Sensitivity-Specificity-Accuracy Curve ({model_name})")
    plt.show()

    print(f"Optimal Threshold for {model_name}: {optimal_thresh:.3f}")

# Integrating SSA Curve into Model Evaluation
def evaluate_model(model, X_resampled, y_resampled, model_name):
    model.fit(X_resampled, y_resampled)
    y_pred_proba = model.predict_proba(X_resampled)[:, 1]
    f1 = f1_score(y_resampled, model.predict(X_resampled))
    roc_auc = roc_auc_score(y_resampled, y_pred_proba)
    plot_ssa_curve(y_resampled, y_pred_proba, model_name)  # Added SSA Curve
    print(f"{model_name} - F1 Score: {f1}, ROC-AUC Score: {roc_auc}")
    return {'F1 Score': f1, 'ROC-AUC': roc_auc}

# Running SSA Curve for all models
eval_results = {}
for name, model in models.items():
    eval_results[name] = evaluate_model(model, X_resampled, y_resampled, name)

# Display summary
eval_df = pd.DataFrame(eval_results).T
print(eval_df)
```
